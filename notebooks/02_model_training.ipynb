# 02 — Model training (SMOTE + RF / XGBoost)

The canonical training script lives at `../scripts/train_model.py` and writes artifacts to `../backend/`.
This notebook mirrors that flow for experimentation.

```python
import sys
from pathlib import Path
ROOT = Path("..").resolve()
sys.path.insert(0, str(ROOT / "backend"))
sys.path.insert(0, str(ROOT / "scripts"))
import pandas as pd
from feature_engineering import add_engineered_columns
import train_model as tm

tm.ensure_data()
df = pd.read_csv(tm.DATA_CSV)
df.columns = [c.strip() for c in df.columns]
for c in ["TWF", "HDF", "PWF", "OSF", "RNF"]:
    if c not in df.columns:
        df[c] = 0
df = add_engineered_columns(df)
df["failure_type"] = df.apply(tm.failure_type, axis=1)
df["failure_type"].value_counts()
```
